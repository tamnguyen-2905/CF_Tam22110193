# Momentum trading

```python
import yfinance as yf
import pandas as pd
import numpy as np

def get_data(tickers, start_date):
    # Láº¥y dá»¯ liá»‡u giÃ¡ Ä‘Ã³ng cá»­a Ä‘iá»u chá»‰nh
    print(f"Äang táº£i dá»¯ liá»‡u cho: {tickers}...")
    data = yf.download(tickers, start=start_date, auto_adjust=False, progress=False)['Adj Close']
    return data

def analyze_market_mood(data, n_stocks=3):
    """
    HÃ m nÃ y chá»‰ phÃ¢n tÃ­ch vÃ  in ra tráº¡ng thÃ¡i, khÃ´ng thá»±c hiá»‡n giao dá»‹ch.
    """
    # 1. Chuyá»ƒn Ä‘á»•i sang dá»¯ liá»‡u cuá»‘i thÃ¡ng
    monthly_prices = data.resample('ME').last()

    # 2. TÃ­nh lá»£i suáº¥t Logarit thÃ¡ng gáº§n nháº¥t
    # CÃ´ng thá»©c: ln(P_t / P_{t-1})
    monthly_log_returns = np.log(monthly_prices / monthly_prices.shift(1))

    # 3. Láº¥y dá»¯ liá»‡u cá»§a thÃ¡ng hoÃ n thiá»‡n gáº§n nháº¥t (DÃ²ng cuá»‘i cÃ¹ng)
    latest_date = monthly_log_returns.index[-1]
    latest_performance = monthly_log_returns.iloc[-1].dropna()

    if latest_performance.empty:
        print("KhÃ´ng Ä‘á»§ dá»¯ liá»‡u Ä‘á»ƒ xáº¿p háº¡ng thÃ¡ng gáº§n nháº¥t.")
        return

    # 4. Xáº¿p háº¡ng (Ranking)
    # Sáº¯p xáº¿p tá»« cao xuá»‘ng tháº¥p
    ranked_stocks = latest_performance.sort_values(ascending=False)

    # 5. Lá»c ra Top Hot vÃ  Top Cold
    hot_stocks = ranked_stocks.head(n_stocks)
    cold_stocks = ranked_stocks.tail(n_stocks)

    # --- IN Káº¾T QUáº¢ ---
    print("\n" + "="*40)
    print(f" BÃO CÃO MOMENTUM THÃNG: {latest_date.strftime('%m-%Y')}")
    print("="*40)

    print(f"\nðŸ”¥ TOP {n_stocks} Cá»” PHIáº¾U TÄ‚NG Máº NH:")
    for ticker, ret in hot_stocks.items():
        # Chuyá»ƒn log return sang % Ä‘Æ¡n giáº£n cho dá»… Ä‘á»c: (e^r - 1) * 100
        pct_change = (np.exp(ret) - 1) * 100
        print(f"   - {ticker}: +{pct_change:.2f}%")

    print(f"\nâ„ï¸ TOP {n_stocks} Cá»” PHIáº¾U Yáº¾U:")
    for ticker, ret in cold_stocks.items():
        pct_change = (np.exp(ret) - 1) * 100
        print(f"   - {ticker}: {pct_change:.2f}%")

    print("\n" + "="*40)

# --- THá»°C THI ---

universe = [
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META', 'TSLA', 'NVDA', 'NFLX', 'INTC', 'AMD'
]

# Láº¥y dá»¯ liá»‡u tá»« quÃ¡ khá»© Ä‘áº¿n hiá»‡n táº¡i
# start_date khÃ´ng cáº§n quÃ¡ xa, chá»‰ cáº§n Ä‘á»§ Ä‘á»ƒ tÃ­nh return thÃ¡ng gáº§n nháº¥t
start_date = '2023-01-01'

df_prices = get_data(universe, start_date)

# In ra 3 mÃ£ máº¡nh nháº¥t vÃ  3 mÃ£ yáº¿u nháº¥t
analyze_market_mood(df_prices, n_stocks=5)
```
